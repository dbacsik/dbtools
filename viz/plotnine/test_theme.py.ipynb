```python
import pandas as pd
import plotnine as p9
from plotnine.data import mpg

from theme_clinical import theme_clinical, palette_clinical
```

```python
p = (p9.ggplot(mpg.dropna(), p9.aes(x='displ', fill='drv'))
     + p9.geom_density(color='black', alpha=0.6)
     + p9.scale_fill_manual(values = palette_clinical)
     + p9.labs(title="Displacement by Drive Type",
             x="Displacement",
             y="Density")
     + theme_clinical()
)

p
```
