```r
library(ggplot2)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
```

```r
sample_dot <- ggplot(mpg, aes(x = displ, y = hwy, colour = drv)) +
  geom_point(size = 2.5) +
  labs(
    title = "Fuel Efficiency by Engine Displacement",
    subtitle = "Data from the US EPA, 1999-2008",
    x = "Engine Displacement (Litres)",
    y = "Highway Fuel Efficiency (MPG)",
    colour = "Vehicle Class",
    caption = "Source: ggplot2::mpg dataset"
  )

sample_dot
```

```r
sample_histo <- ggplot(mpg, aes(x = displ, fill = drv)) +
  geom_histogram(binwidth=0.25, alpha=0.8) +
  labs(
    title = "Fuel Efficiency by Engine Displacement",
    subtitle = "Data from the US EPA, 1999-2008",
    x = "Engine Displacement (Litres)",
    y = "Highway Fuel Efficiency (MPG)",
    caption = "Source: ggplot2::mpg dataset"
  )

sample_histo
```
