```python
import plotnine as p9
from plotnine.data import mpg
```

```python
p9.options.figure_size = (10, 5)
```

```python
sample_dot = (p9.ggplot(mpg, p9.aes(x='displ', y='hwy', colour='drv'))
  + p9.geom_point(size=2.5)
  + p9.labs(
    title="Fuel Efficiency by Engine Displacement",
    subtitle="Data from the US EPA, 1999-2008",
    x="Engine Displacement (Litres)",
    y="Highway Fuel Efficiency (MPG)",
    colour="Vehicle Class",
    caption="Source: plotnine.data::mpg dataset"
  )
)
sample_dot
```

```python
sample_histo = (p9.ggplot(mpg, p9.aes(x='displ', fill='drv'))
  + p9.geom_histogram(binwidth=0.25, alpha=0.8)
  + p9.labs(
    title="Fuel Efficiency by Engine Displacement",
    subtitle="Data from the US EPA, 1999-2008",
    x="Engine Displacement (Litres)",
    y="Count",
    fill="Drive type",
    caption="Source: plotnine.data::mpg dataset"
  )
)
sample_histo
```
